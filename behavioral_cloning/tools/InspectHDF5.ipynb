```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def display_excerpt(file_path, n=8):
    with h5py.File(file_path) as f:
        data = f['data']
        label = f['label']
    
        plt.figure(figsize=(10, n * 1.5))
        for i, idx in enumerate(np.random.choice(label.shape[0], n)):
            #print(data[idx, :, :, :].shape)
            plt.subplot(n/2, 2, i + 1)
            plt.imshow(np.asarray(data[idx, :, :, :].transpose([1, 2, 0]).squeeze() * 255, dtype=np.uint8), cmap='gray')
            plt.annotate(str(label[idx]), (150, 150), color='white')
            plt.xticks([])
            plt.yticks([])
        plt.show()
        
```
